```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import xlrd
import xlwt
import xlsxwriter
```

```python
excel_sheets = pd.ExcelFile('EM_Risk_Monitor.xlsx')

sheet_names = pd.read_excel('EM_Risk_Monitor.xlsx', sheet_name=None)

sheets = list(sheet_names.keys())

for i in range(len(sheets)):
    sheets[i] = pd.read_excel('EM_Risk_Monitor.xlsx', sheet_name=sheets[i])
```

Read information from an excel sheet

```python
names = {k: None for k in list(sheet_names.keys())}
```

```python
final_DF = dict(zip(names, sheets))
```

We now assume that there aren't missing columns and stuff - if data is missing, we impute.

```python
#FIll in missing values by imputing the data
```

```python
for items in names:
    final_DF[items].fillna(final_DF[items].mean(), inplace=True)
```

```python
final_DF['3_Month_Deposit_Rate']
```

mAKE THE Timeline column a datetime

```python
for sheet_nam in names:
    final_DF[sheet_nam].timeline = pd.to_datetime(final_DF[sheet_nam].timeline)
    final_DF[sheet_nam].set_index('timeline', inplace=True)
```

I am going to attempt a multiple regression of unnormalized and unstandardized data

```python
#Check whether a relationship exist
#Scatter plots are a helpful tool for that

plt.scatter(final_DF['fiveYearCDS'], final_DF['3_Month_Deposit_Rate'], color='red')
plt.show()
```
